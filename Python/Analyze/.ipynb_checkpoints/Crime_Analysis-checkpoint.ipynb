# 목표
경기도 내 마약범죄 시각화<br>맨 아래 내용 확인가능

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.rcParams['font.family'] = 'AppleGothic'

# 유니코드 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
file_path = '/Users/yoominsang/study/Download/Data/crime_20231231.csv'
```

```python
df = pd.read_csv(file_path, encoding='euc-kr')
```

```python
print(df.head(38))
```

```python
# 외국 데이터 제거
df_drop_foreigner = df.drop(columns=df.columns[df.columns.str.startswith('외국')])

for col in df_drop_foreigner:
    print(col)
```

```python
df_gyeonggi_row = df.columns[df.columns.str.startswith('경기도')]

for col in df_gyeonggi_row:
    print(col)
```

```python
print(df_gyeonggi.shape)
```

```python
df_gyeonggi = df.loc[:,df.columns.str.contains('범죄대분류|범죄중분류|경기도')]

for col in df_gyeonggi:
    print(df_gyeonggi.head(38))
```

```python
# 경기도 마약 범죄만 조회
df_drug_crime = df_gyeonggi[df_gyeonggi['범죄중분류'] == '마약범죄']

print(df_drug_crime.head())
```

```python
reg_city_cols = [col for col in df_drug_crime.columns if col.startswith('경기도')]

df_long = df_drug_crime.melt(
    id_vars=['범죄대분류', '범죄중분류'],
    value_vars=reg_city_cols, # 경기도 도시 컬럼들만 선택
    var_name='도시',
    value_name='발생건수'
)

print(df_long.head())
```

```python
# 막대그래프
plt.figure(figsize=(12,7))

sns.barplot(data=df_long, x='도시', y='발생건수', palette='viridis')

# 그래프 제목 및 축 레이블 설정
plt.title('경기도 내 시별 마약범죄 발생 건수', fontsize=16)
plt.xlabel('도시', fontsize=12)
plt.ylabel('마약범죄 발생 건수', fontsize=12)

# Y축 레이블의 숫자를 정수 형태로 포맷 (선택 사항)
plt.ticklabel_format(style='plain', axis='y')

# x축 레이블이 길 경우 겹치지 않도록 회전
plt.xticks(rotation=90, ha='right')

# 레이아웃 자동 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
